#### Question 2

User Rating of a restaurant plays a crucial role in selecting a restaurant or ordering the food from the restaurant.

    1. Write a short detail analysis of how the rating is affected by restaurant due following features: Plot a suitable graph to explain your inference.
        1. Number of Votes given Restaurant
        2. Restaurant serving more number of cuisines.
        3. Average Cost of Restaurant
        4. Restaurant serving some specific cuisines.
    2. Find the weighted restaurant rating of each locality and find out the top 10 localities with more weighted restaurant rating?
        1. Weighted Restaurant Rating=Σ (number of votes * rating) / Σ (number of votes) .


```python
import requests
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict

zomato = pd.read_csv('E:\Coding Ninjas Data Science\Dataset\zomato.csv', encoding='latin-1')
df = zomato.copy()

plt.figure(figsize=(10, 10))
plt.scatter(df['Aggregate rating'], df['Votes'], alpha=0.5, s=20, color='blue', edgecolor='black')
plt.yticks(np.arange(0, 12000, 1000))
plt.axis([0, 5, 0, 12000])
plt.xlabel('Rating --->', size=12)
plt.ylabel('Number of votes --->', size=12)
plt.title('Rating v/s Number of Votes')
plt.grid()
plt.show()
```

#### Analysis:

It can be inferred from the graph that as the number of votes increases the chances of getting higher ratings also increases i.e. it appears to be directly proportional. Any restautrant with more than 5000 votes has been rated between 4 and 5. For lower number of votes i.e. where the votes are less than 2000, the ratings varies over a wide range from 2-5. Therefore, for a restaurant to have a probability of getting higher aggregate rating, it should have more than higher number of votes. Additionally, there are many restaurants having votes between between 1000 and 2000 with very good ratings i.e. between 4 and 5.

```python
df=df.loc[df.Cuisines.notnull()]
cuisines = [len(cuisine.split(',')) for cuisine in df.Cuisines]

plt.figure(figsize=(10, 10))
plt.scatter(df['Aggregate rating'], cuisines, color='blue', edgecolor='black', alpha=0.5, s=20)
plt.xlabel('Rating --->', size=12)
plt.ylabel('Number of cuisines --->', size=12)
plt.title('Rating v/s Number of Cuisines')
plt.grid()
plt.show()
```

#### Analysis:

It can be inferred from the graph that when the number of cuisines provided increases from 3 to 8, generally the rating seems to converge between 3 and 4. Restaurants providing more number of cuisines are not much likely to get higher ratings, especially when the number of cuisines provided exceeds 6. While if the restaurants provide less number of cuisines, for example 1 or 2 cuisines, they are more likely to get higher ratings. In fact many restaurants in India are providing 2 cuisines. There is a very less number of restaurants providing 8 cuisines. It seems like when a restaurant provides too many cuisines, its focus on the quality of food offered diverges. While restaurants providing less cuisines focus on the quality of food to get good aggregate ratings.

```python
df = zomato.copy()

plt.figure(figsize=(10,10))
plt.scatter(df['Aggregate rating'], df['Average Cost for two'], s=20, color='blue', alpha=0.6)
plt.xlabel('Aggregate Rating --->', size=12)
plt.ylabel('Average Cost for two--->', size=12)
plt.title('Aggregate rating versus average cost')
plt.grid()
plt.show()

plt.figure(figsize=(10,10))
plt.scatter(df['Aggregate rating'], df['Average Cost for two'], s=15, c='orange', alpha=0.6)
plt.grid()
plt.axis([0, 5, 0, 9000])
plt.xlabel('Aggregate Rating --->', size=12)
plt.ylabel('Average Cost for two --->', size=12)
plt.title('Aggregate rating versus average cost')
plt.show()
```

#### Analysis:


In the first graph as average cost goes higher than 1 lakh the rating tries to consolidate between 4-5, which demonstrates that the food quality must be good. This graph doesn’t offer much details about as much of the graph is consolidated at the lower levels on y-axis. Therefore, plotted the second graph to by narrowing the average cost to 9000 on y-axis. It can be inferred from the second graph that lower the average cost, more dispersed is the food quality as the rating goes from 2-5. However, as the average cost increases the aggregate rating consolidates in the upper levels which means the food quality becomes better at that price point.


```python
df = zomato.copy()
df = df.loc[df.Cuisines.notnull()]

cuisines = []

for cuisine in df.Cuisines:
    for cus in cuisine.split(','):
        cuisines.append(cus.strip())
        
cuisines_count = defaultdict(int)

for cuisine in cuisines:
    cuisines_count[cuisine] += 1
    
xaxis = [key for key in cuisines_count.keys()]
yaxis = [value for value in cuisines_count.values()]

np_xaxis = np.array(xaxis)
np_yaxis = np.array(yaxis)

np_xaxis = np_xaxis[np.argsort(np_yaxis)]
np_yaxis = np.sort(np_yaxis)

np_xaxis = np_xaxis[::-1][:5]
np_yaxis = np_yaxis[::-1][:5]

plt.figure(figsize=(10, 10))
for cuisine in np_xaxis:
    rating = []
    for cus, rat in zip(df.Cuisines, df['Aggregate rating']):
        if cuisine in cus:
            rating.append(rat)
    plt.hist(rating, edgecolor='black', bins=[0, 1, 2, 3, 4, 5])
    plt.xlabel('Rating --->')
    plt.ylabel('Number of restaurants --->')
    plt.title('Rating v/s Specific Cuisine')
plt.grid()
plt.legend(labels=np_xaxis)
plt.show()
```

#### Analysis:

It can be inferred from the above graph that North Indian cuisine is the best rated among all the cuisines in all aggregate ranges. After that Chinese cuisine follows, as for the maximum number of times it has a rating of either 3 or 4. Next follows Fast Food which as also has a rating of either 3 or 4 maximum number of times. These are then followed by Mughlai and Italian cuisines.

```python
df = zomato.copy()

rating = np.array([r for r in df['Aggregate rating']])
votes = np.array([v for v in df['Votes']])
locality = np.array([l for l in df['Locality']])

locality_dict = defaultdict(float)
for local in locality:
    if sum(votes[np.argwhere(locality==local)])[0]!=0:
        locality_dict[local] = sum((rating*votes)[np.argwhere(locality==local)])[0]/sum(votes[np.argwhere(locality==local)])[0]
    
np_locality = np.array([key for key in locality_dict])
np_weight = np.array([value for value in locality_dict.values()])

np_locality = np_locality[np.argsort(np_weight)]
np_weight = np.sort(np_weight)

np_locality = np_locality[::-1]
np_weight = np_weight[::-1]

for i in range(len(np_locality)):
    print(np_locality[i], '--->', np_weight[i])

np_locality = np_locality[:10]
np_weight = np_weight[:10]

plt.figure(figsize=(10,10))
plt.bar(np_locality, np_weight, color='orange', edgecolor='black', width=0.5)
plt.xticks(rotation='vertical')
plt.title('Locality v/s Weighted Rating (Top 10)')
plt.xlabel('Localities --->')
plt.ylabel('Weighted Rating --->')
plt.show()

for i in range(len(np_locality)):
    print(np_locality[i], '--->', np_weight[i])
```
