Visualization

    1. Plot the bar graph top 15 restaurants have a maximum number of outlets.
    2. Plot the histogram of aggregate rating of restaurant( drop the unrated restaurant).
    3. Plot the bar graph top 10 restaurants in the data with the highest number of votes.
    4. Plot the pie graph of top 10 cuisines present in restaurants in the USA.
    5. Plot the bubble graph of a number of Restaurants present in the city of India and keeping the weighted restaurant rating of the city in a bubble.


```python
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from collections import defaultdict

zomato = pd.read_csv('E:\Coding Ninjas Data Science\Dataset\zomato.csv', encoding='latin-1')
df = zomato.copy()

restaurants = defaultdict(int)

for name in df['Restaurant Name']:
    restaurants[name] += 1
    
xaxis = []
yaxis = []

for key, value in restaurants.items():
    xaxis.append(key)
    yaxis.append(value)
    
np_xaxis = np.array(xaxis)
np_yaxis = np.array(yaxis)

np_xaxis = np_xaxis[np.argsort(np_yaxis)]
np_yaxis = np.sort(np_yaxis)

np_xaxis = np_xaxis[::-1][:15]
np_yaxis = np_yaxis[::-1][:15]

plt.figure(figsize=(10,10))
plt.bar(np_xaxis, np_yaxis, color='blue', edgecolor='black', width=0.5)
plt.xticks(rotation='vertical')
plt.xlabel('Restaurants Names --->')
plt.ylabel('Number of Outlets --->')
plt.title('Top 15 restaurants with maximum outlets')
plt.grid()
plt.show()

for i in range(len(np_xaxis)):
    print(np_xaxis[i],'--->', np_yaxis[i])
```

#### Analysis:

It can be inferred from the above graph that, Cafe Coffee Day has most number of outlets i.e. 83. It is followed by Domino's Pizza, Subway, Green Chick Chop, McDonald's with 79, 63, 51 and 48 outlets respectively making them the top 5 restaurants. Rest are shown above.

```python
rating = df['Aggregate rating'].dropna()
plt.figure(figsize=(10, 10))
plt.xlabel('Aggregate Rating --->', size=12)
plt.ylabel('Number of restaurants --->', size=12)
plt.hist(rating, color='blue', edgecolor='black', bins=range(6))
plt.title('Histogram for Aggregate Ratings', size=12)
plt.grid()
plt.show()

```

#### Analysis:

It can be inferred from the above histogram that, most number of restaurants have the aggregate rating between 3 and 4. This if followed by the range 0-1 at the second place. Next are the ranges 2-3 and 4-5 with nearly equal number of restaurants, while the range 1-2 has the minimum number of restaurants.

```python
np_restaurants = np.array([r for r in df['Restaurant Name']])
np_votes = np.array([v for v in df['Votes']])

vote_count = defaultdict(int)

for i in range(len(np_votes)):
    vote_count[np_restaurants[i]] += np_votes[i]

xaxis = []
yaxis = []

for key, value in vote_count.items():
    xaxis.append(key)
    yaxis.append(value)
    
np_xaxis = np.array(xaxis)
np_yaxis = np.array(yaxis)

np_xaxis = np_xaxis[np.argsort(np_yaxis)]
np_yaxis = np.sort(np_yaxis)

np_xaxis = np_xaxis[::-1][:10]
np_yaxis = np_yaxis[::-1][:10]

plt.figure(figsize=(10,10))
plt.bar(np_xaxis, np_yaxis, color='blue', edgecolor='black', width=0.5)
plt.xlabel('Restaurants Names --->', size=12)
plt.ylabel('Number of Votes --->', size=12)
plt.title('Top 10 restaurants with highet number of votes', size=12)
plt.xticks(rotation='vertical')
plt.grid()
plt.show()

for i in range(len(np_xaxis)):
    print(np_xaxis[i], '--->', np_yaxis[i])
```

#### Analysis:

It can be inferred from the above that Barbeque Nation has received the most number of votes i.e. 28142 and is miles ahead from AB's - Absolute Barbecues having 13400 votes and holds second position. These are then followed by Toit, Big Chill, Farzi Cafe, Truffles, Chili's, Hauz Khas Social, Joey's Pizza and Peter Cat respectively to make the top 10 list.

```python
df = zomato.copy()

df = df[df['Cuisines'].notnull()]
df = df[df['Country Code'] == 216]

cuisines = []

for c in df['Cuisines']:
    for item in c.split(','):
        cuisines.append(item.strip())

cuisine_count = defaultdict(int)

for cuisine in cuisines:
    cuisine_count[cuisine] += 1

xaxis = []
yaxis = []

for key, value in cuisine_count.items():
    xaxis.append(key)
    yaxis.append(value)
    
np_xaxis = np.array(xaxis)
np_yaxis = np.array(yaxis)

np_xaxis = np_xaxis[np.argsort(np_yaxis)]
np_yaxis = np.sort(np_yaxis)

np_xaxis = np_xaxis[::-1][:10]
np_yaxis = np_yaxis[::-1][:10]

plt.figure(figsize=(10,10))
plt.pie(np_yaxis, labels = np_xaxis, autopct="%.2f%%")
plt.title('Top 10 Cuisines in United States', size=12)
plt.show()

for i in range(len(np_xaxis)):
    print(np_xaxis[i], '--->', np_yaxis[i])
```

#### Analysis:

The above pie chart shows the top 10 cuisines severed in United States and their share by percentage. It is evident from the above that American cuisine has the largest share with 22.18% which is miles ahead of Seafood with 11.68% holding the second position. These are followed by Pizza, Burger, Sandwich, Steak, Italian, Breakfast, Mexican and Sushi to make the top 10 list.

```python
df = zomato.copy()

df = df[df['Country Code'] == 1]

np_votes = np.array([v for v in df['Votes']])
np_rating = np.array([r for r in df['Aggregate rating']])
np_city = np.array([l for l in df['City']])

weighted_city = defaultdict(int)

for city in set(np_city):
    if sum(np_votes[np.argwhere(np_city == city)])[0] != 0:
        weighted_city[city] = sum((np_votes*np_rating)[np.argwhere(np_city==city)])[0]/sum(np_votes[np.argwhere(np_city == city)])[0]

city_count = defaultdict(int)

for city in np_city:
    city_count[city] += 1
    
np_xaxis = np.array([key for key in city_count.keys()])
np_yaxis = np.array([value for value in city_count.values()])
np_size = np.array([value for value in weighted_city.values()])


plt.figure(figsize=(10,10))
plt.scatter(np_yaxis, np_xaxis, s=5**np_size, color='orange', alpha=0.5, edgecolor='black')
plt.xlabel('Cities --->', size=12)
plt.ylabel('Number of Restaurants --->', size=12)
plt.title('Number of Restaurants in a City with weighted rating as bubble size')
plt.xticks(rotation='vertical')
plt.grid()
plt.show()


plt.figure(figsize=(10,10))
plt.scatter(np_yaxis, np_xaxis, s=5**np_size, color='red', alpha=0.5, edgecolor='black')
plt.xlabel('Cities --->', size=12)
plt.ylabel('Number of Restaurants --->', size=12)
plt.title('Number of Restaurants in a City with weighted rating as bubble size')
plt.xticks(rotation='vertical')
plt.grid()
plt.axis([0,50,'Agra', 'Vizag'])
plt.show()
```

####  Analysis:

Two graphs have been plotted above to better analyse the details. From the first graph, it is evident that New Delhi has the maximum number of restaurants which is miles ahead from any other cities depicted in the graph but it New Delhi does not appear to have the highest weighted average as bigger bubbles are clearly seen in the first graph. It is also visible from the graph that, for most of the cities, the bubble are plotted around the 0 on x-axis. To better analyse this, second graph has been plotted cropping the x-axis to between 0 and 50. From the second graph it is clear that, for most cities the number of restaurants are consolidated around 20 with Mangalore, Allahabad and Vadodara having visibly the larger bubbles indicating the higher weighted average.
