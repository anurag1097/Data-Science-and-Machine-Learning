```python
# importing all the required libraries.

import time
import re
from selenium import webdriver
from bs4 import BeautifulSoup as bs
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException
```

```python
# Defining different functions for performing certain actions to avoid repititive code as they care called multiple times.
# All the functions below are documented individually.

def getChromeWindow(url):
    """
    This function is used to open the chrome window and go to the desired website.
    :param url: URL of the website you want to open.
    :return driver: Driver handle of the chrome window.
    """
    driver = webdriver.Chrome(executable_path = 'E:\Coding Ninjas Data Science\Lecture 20 - Web Browsing Selenium\chromedriver.exe')
    driver.get(url) # url - 'https://www.instagram.com/'
    driver.maximize_window()
    return driver
    
def loginToInstagram(driver, username, password):
    """
    This function is used to login into the desired instagram account and to avoid notifications.
    :param username: Username of the Instagram account .
    :param password: Password of the Instagram account.
    :param driver: Driver of the Instagram handle.
    """
    username_box = driver.find_element_by_name('username')
    password_box = driver.find_element_by_name('password')
    username_box.send_keys(username)
    password_box.send_keys(password)
    login = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[@class = "sqdOP  L3NKy   y3zKF     "]')))
    login.submit()
    login_info = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class = 'cmbtv']/button")))
    login_info.click()
    notification = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'aOOlW   HoLwm ']")))
    notification.click()
        
def searchAndPrintDropDownList(driver, keyword):
    """
    This function is used to print the names top accounts from the drop down after typing the word in the search box.
    :param keyword: Word to be used.
    :param driver: Driver of the chrome window.
    """
    search = driver.find_element_by_xpath('//input[@placeholder = "Search"]')
    search.send_keys(keyword)
    waiter = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.CLASS_NAME, "Ap253")))
    drp_dwn = driver.find_elements_by_class_name('yCE8d  ')
    for account in drp_dwn:
        if account.get_attribute('href').find('https://www.instagram.com/explore/locations/') == -1 and account.get_attribute('href').find('https://www.instagram.com/explore/tags') == -1:
            name = account.find_element_by_class_name('Ap253').text
            print(name)
            
def searchAndOpenProfile(driver, keyword, acc_href):
    """
    This function is used to search for a profile using the search box and opening the account.
    :param keyword: Word to be searched in the search box.
    :param acc_href: This is the hyperlink to the desired account. This is generally same as the username of the account.
    :param driver: Driver of the chrome window.
    """
    try:
        clear_btn = driver.find_element_by_class_name('aIYm8')
        clear_btn.click()
        search = driver.find_element_by_xpath('//input[@placeholder = "Search"]')
        search.send_keys(keyword)
        acc = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@href="/{}/"]'.format(acc_href))))
        acc.click()
    except NoSuchElementException:
        search = driver.find_element_by_xpath('//input[@placeholder = "Search"]')
        search.send_keys(keyword)
        wait = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class = 'drKGC']")))
        acc = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@href="/{}/"]'.format(acc_href))))
        acc.click()
    
def followAndUnfollowAccount(driver, keyword, acc_href, mode):
    """
    This method is used to follow or unfollow any account.
    :param driver: Driver of the chrome window.
    :param keyword: Word to be searched in the search box.
    :param acc_href: This is the hyperlink to the desired account. This is generally same as the username of the account.
    :param mode: Two valid option - follow/unfollow. This is used to define whether you want to follow or unfollow any account.
    """
    if acc_href != driver.find_element_by_tag_name('h2').text:
            searchAndOpenProfile(driver, keyword, acc_href)
    if mode == 'follow':
        try:
            if driver.find_element_by_xpath('//button[contains(@class, "sqdOP")]').get_attribute('innerHTML') == 'Message':
                print("You're already following this account")
            else:
                raise NoSuchElementException
        except NoSuchElementException:
            follow_btn = driver.find_element_by_class_name('_5f5mN')
            follow_btn.click()
    elif mode == 'unfollow':
        try:
            if driver.find_element_by_xpath('//button[contains(@class, "sqdOP")]').get_attribute('innerHTML') == 'Message':
                unfollow_btn = driver.find_element_by_class_name('_5f5mN')
                unfollow_btn.click()
                unfollow_confirmation = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, "//button[contains(@class,'-Cab_')]")))
                unfollow_confirmation.click()
            else:
                raise NoSuchElementException
        except NoSuchElementException:
            print("You've already unfollowed this account")
            
def likeAndUnlikePosts(driver, keyword, acc_href, num, mode):
    """
    This method is used to like or unlike posts for any account.
    :param driver: Driver of the chrome window.
    :param keyword: Word to be searched in the search box.
    :param acc_href: This is the hyperlink to the desired account. This is generally same as the username of the account.
    :param mode: Two valid option - like/unlike. This is used to define whether you want to like or unlike the posts.
    :param num: This is used to define the number of posts you want to like or unlike.
    """
    if acc_href != driver.find_element_by_tag_name('h2').text:
            searchAndOpenProfile(driver, keyword, acc_href)
    time.sleep(2)
    if mode == 'like':
        first_post = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class = 'v1Nh3 kIKUG  _bz0w']")))
        first_post.click()
        like_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'wpO6b ']/div/span")))
        bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
        if bs_data.find('svg')['aria-label'] == 'Like':
            like_button.click()
        else:
            print("You've already like this picture")
        count = 1
        while count < num:
            next_btn = driver.find_element_by_xpath("//a[@class = ' _65Bje  coreSpriteRightPaginationArrow']")
            next_btn.click()
            like_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'wpO6b ']/div/span")))
            bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
            if bs_data.find('svg')['aria-label'] == 'Like':
                like_button.click()
                time.sleep(0.8)
            else:
                print("You've already like this picture")
            count += 1
            if count == num:
                print("You've liked all {} pictures".format(num))
                
    elif mode == 'unlike':
        first_post = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//div[@class = 'v1Nh3 kIKUG  _bz0w']")))
        first_post.click()
        like_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'wpO6b ']/div/span")))
        bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
        if bs_data.find('svg')['aria-label'] == 'Unlike':
            like_button.click()
        else:
            print("You've already unliked this post")
        count = 1
        while count < num:
            next_btn = driver.find_element_by_xpath("//a[@class = ' _65Bje  coreSpriteRightPaginationArrow']")
            next_btn.click()
            like_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'wpO6b ']/div/span")))
            bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
            if bs_data.find('svg')['aria-label'] == 'Unlike':
                like_button.click()
            else:
                print("You've already unliked this post")
            count += 1
            if count == num:
                print("You've unliked all {} posts".format(num))
    li = driver.find_elements_by_class_name('wpO6b ')
    close_btn = li[len(li)-1]
    close_btn.click()
    
def extractFollowers(driver, keyword, acc_href, num):
    """
    This method is used etract the list of followers for any account. This clicks on the followers button and scroll down to extract.
    :param driver: Driver of the chrome window.
    :param keyword: Word to be searched in the search box.
    :param acc_href: This is the hyperlink to the desired account. This is generally same as the username of the account.
    :param num: This defines the number of followers to be extracted. For my code below, as per TA's suggestion this has been
                limited to 500 as otherwise the code will run for very long time.
    """
    if acc_href != driver.find_element_by_tag_name('h2').text:
            searchAndOpenProfile(driver, keyword, acc_href)
    time.sleep(2)
    followers_btn = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//a[@class = "-nal3 "]/span')))
    followers_btn.click()
    wait = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
    while len(driver.find_elements_by_xpath('//a[@class = "FPmhX notranslate  _0imsa "]')) < num:
        driver.execute_script('arguments[0].scrollIntoView(0, 1000);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
    followers = driver.find_elements_by_xpath('//a[@class = "FPmhX notranslate  _0imsa "]')
    followers = followers[:500]
    final = []
    for follower in followers:
        print(follower.text)
        final.append(follower.text)
    li = driver.find_elements_by_class_name('wpO6b ')
    close_btn = li[len(li)-1]
    close_btn.click()
    return final

def extractFollowing(driver, keyword, acc_href, num):
    """
    This method is used etract the list of following for any account. This clicks on the following button and scroll down to extract.
    :param driver: Driver of the chrome window.
    :param keyword: Word to be searched in the search box.
    :param acc_href: This is the hyperlink to the desired account. This is generally same as the username of the account.
    :param num: This defines the number of following's to be extracted. For my code below, as per TA's suggestion this has been
                limited to 500 as otherwise the code will run for very long time.
    """
    if acc_href != driver.find_element_by_tag_name('h2').text:
            searchAndOpenProfile(driver, keyword, acc_href)
    time.sleep(2)
    following_btn = driver.find_elements_by_xpath('//a[@class="-nal3 "]/span[@class="g47SY "]')[1]
    following_btn.click()
    wait = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//ul[contains(@class, "jSC57")]')))
    while len(driver.find_elements_by_xpath('//a[@class = "FPmhX notranslate  _0imsa "]')) < num:
        driver.execute_script('arguments[0].scrollIntoView(0, 1000);', driver.find_element_by_xpath('//ul[contains(@class, "jSC57")]'))
    following = driver.find_elements_by_xpath('//a[@class = "FPmhX notranslate  _0imsa "]')
    following = following[:500]
    final = []
    for followed in following:
        print(followed.text)
        final.append(followed.text)
    li = driver.find_elements_by_class_name('wpO6b ')
    close_btn = li[len(li)-1]
    close_btn.click()
    return final

def checkUserStory(driver, keyword, acc_href):
    """
    This method is used to check the Instagram story for any account.
    :param driver: Driver of the chrome window.
    :param keyword: Word to be searched in the search box.
    :param acc_href: This is the hyperlink to the desired account. This is generally same as the username of the account.
    """
    if acc_href != driver.find_element_by_tag_name('h2').text:
            searchAndOpenProfile(driver, keyword, acc_href)
    time.sleep(2)
    try:
        if int(driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]/canvas').get_attribute('height'))==168:
            print('You have not seen the story yet. Opening the story!')
            driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]').click()
        elif int(driver.find_element_by_xpath('//div[contains(@class, "h5uC0")]/canvas').get_attribute('height'))==166:
            print('You have already seen the story!')
    except NoSuchElementException:
        print('{} does not have any story as of now!'.format(keyword))
```
