### Question 1

1. Login to your Instagram Handle

    1. Submit with sample username and password


```python
import time
import re
from selenium import webdriver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.common.exceptions import StaleElementReferenceException
```

```python
def getChromeWindow(url):
    driver = webdriver.Chrome(executable_path = 'E:\Coding Ninjas Data Science\Lecture 20 - Web Browsing Selenium\chromedriver.exe')
    driver.get(url) # url - 'https://www.instagram.com/'
    driver.maximize_window()
    return driver
    
def loginToInstagram(driver, username, password):
    username_box = driver.find_element_by_name('username')
    password_box = driver.find_element_by_name('password')
    username_box.send_keys(username)
    password_box.send_keys(password)
    login = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, '//button[@class = "sqdOP  L3NKy   y3zKF     "]')))
    login.submit()
    time.sleep(2)
    login_info = driver.find_element_by_xpath("//button[@class = 'sqdOP yWX7d    y3zKF     ']")
    login_info.click()
    time.sleep(2)
    notification = driver.find_element_by_xpath("//button[@class = 'aOOlW   HoLwm ']")
    notification.click()
        
```

```python
driver = getChromeWindow('https://www.instagram.com/')
loginToInstagram(driver, 'anu_97_', 'Instagram@1997')
```

### Question 4

4. Follow/Unfollow given handle - 

    1. Open the Instagram Handle of “So Delhi”
    2. Start following it. Print a message if you are already following
    3. After following, unfollow the instagram handle. Print a message if you have already unfollowed.


```python


try:
    if driver.find_element_by_xpath('//button[contains(@class, "sqdOP")]').get_attribute('innerHTML') == 'Message':
        print("You're already following this account")
    else:
        raise NoSuchElementException
except NoSuchElementException:
    follow_btn = driver.find_element_by_class_name('_5f5mN')
    follow_btn.click()
```

```python

try:
    if driver.find_element_by_xpath('//button[contains(@class, "sqdOP")]').get_attribute('innerHTML') == 'Message':
        unfollow_btn = driver.find_element_by_class_name('_5f5mN')
        unfollow_btn.click()
        unfollow_confirmation = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, "//button[contains(@class,'-Cab_')]")))
        unfollow_confirmation.click()
    else:
        raise NoSuchElementException
except NoSuchElementException:
    print("You've already unfollowed this account")
```

### Question 5

5. Like/Unlike posts

    1. Liking the top 30 posts of the ‘dilsefoodie'. Print message if you have already liked it.
    2. Unliking the top 30 posts of the ‘dilsefoodie’. Print message if you have already unliked it.


```python
driver.back()
```

```python


try:
    search.send_keys('dilsefoodie')
except StaleElementReferenceException:
    search = driver.find_element_by_xpath('//input[@placeholder = "Search"]')
    search.send_keys('dilsefoodie')
    
dilsefoodie_acc = WebDriverWait(driver, 3).until(EC.presence_of_element_located((By.XPATH, '//a[@href="/dilsefoodie/"]')))
dilsefoodie_acc.click()
```

```python
from bs4 import BeautifulSoup as bs
first_post = driver.find_element_by_xpath("//div[@class = 'v1Nh3 kIKUG  _bz0w']")
first_post.click()
like_button = driver.find_element_by_xpath("//button[@class = 'wpO6b ']/div/span")
bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
if bs_data.find('svg')['aria-label'] == 'Like':
    like_button.click()
else:
    print("You've already like this picture")
count = 1
while count<30:
    next_btn = driver.find_element_by_xpath("//a[@class = ' _65Bje  coreSpriteRightPaginationArrow']")
    next_btn.click()
    like_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'wpO6b ']/div/span")))
    bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
    if bs_data.find('svg')['aria-label'] == 'Like':
        like_button.click()
    else:
        print("You've already like this picture")
    count += 1
    if count == 30:
        print("You've liked all 30 pictures")
```

```python
count = 0

first_post = driver.find_element_by_xpath("//div[@class = 'v1Nh3 kIKUG  _bz0w']")
first_post.click()
like_button = driver.find_element_by_xpath("//button[@class = 'wpO6b ']/div/span")
bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
if bs_data.find('svg')['aria-label'] == 'Unlike':
    like_button.click()
else:
    print("You've already unliked this post")
count = 1
while count<30:
    next_btn = driver.find_element_by_xpath("//a[@class = ' _65Bje  coreSpriteRightPaginationArrow']")
    next_btn.click()
    like_button = WebDriverWait(driver, 10).until(EC.presence_of_element_located((By.XPATH, "//button[@class = 'wpO6b ']/div/span")))
    bs_data = bs(like_button.get_attribute('innerHTML'), 'html.parser')
    if bs_data.find('svg')['aria-label'] == 'Unlike':
        like_button.click()
    else:
        print("You've already unliked this post")
    count += 1
    if count == 30:
        print("You've unliked all 30 post")
```
